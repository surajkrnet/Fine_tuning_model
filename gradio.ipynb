```python
!pip install gradio_client gradio -q

import gradio as gr
from gradio_client import Client, handle_file

# ── PASTE YOUR HF TOKEN HERE ──────────────────────────────────
HF_TOKEN = "*****"
# ─────────────────────────────────────────────────────────────

client = Client("fancyfeast/joy-caption-alpha-two", hf_token=HF_TOKEN)

def caption(image_path, caption_type):
    return client.predict(
        input_image=handle_file(image_path),
        caption_type=caption_type,
        caption_length="any",
        extra_options=[],
        name_input="",
        custom_prompt="",
        api_name="/stream_chat"
    )

demo = gr.Interface(
    fn=caption,
    inputs=[
        gr.Image(type="filepath"),
        gr.Radio(choices=["Descriptive", "Booru tag list"], value="Descriptive"),
    ],
    outputs=gr.Textbox(lines=6),
    title="JoyCaption — Description & Tags",
)

demo.launch(share=True)
```

```python
# weburl
!pip install gradio transformers -q

import gradio as gr
from transformers import pipeline

clf = pipeline("sentiment-analysis")

def analyze(text):
    result = clf(text)[0]
    return result["label"], round(result["score"], 3)

demo = gr.Interface(
    fn=analyze,
    inputs=gr.Textbox(label="Type any sentence"),
    outputs=[gr.Label(label="Sentiment"), gr.Number(label="Confidence")],
    title="Sentiment Analyser",
    examples=[["I love this product!"], ["This is terrible."], ["It was okay."]]
)

demo.launch(share=True)
```
